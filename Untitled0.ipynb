#Pasos para calcular dla mediana


1.   Ordenar los datos de menor a mayor
2.   Calcular la posición de la mediana $k=\frac{1}{2}(n+1)$
3. Calcular la mediana dependiendo

Ejemplo: Para los siguientes datos calcular la medana

$$A=\{4, 5, 3, 7, 9, 8, 7\}$$

#Pasos:

1. Ordenar el conjunto $A$
$$A=\{3, 4, 5, 7, 7, 8, 9\}$$
2. Calcular la posición de la mediana:
$$k=\frac{1}{2}(7+1)=4$$

3. Calcular la mediana. La mediana es el dato de la posición $x_4$.
$$x_4=7$$






```python
m=[4,5,3,7,8,8,7]
m_1= sorted(m)
p=len(m)
k=(1/2)*(p+1)
print("la posición es ",k)
print("la lista ordenada ", m_1)
print("la mediana es " ,m_1[4])
```
